```python
# want year, country, numeric columns in new df
col_df = csv_df[["COUNTRY (DISPLAY)","YEAR (CODE)","Numeric"]]
col_df.head()

# give columns better titles
re = col_df.rename(columns={"COUNTRY (DISPLAY)":"Country", "YEAR (CODE)":"Year", "Numeric":"Gross National Income Per Capita"})
re.head()

# format
re["Gross National Income Per Capita"] = re["Gross National Income Per Capita"].map("${:,.2f}".format)
re.head()

# order by year from highest to lowest
gnipc = re.sort_values("Year", ascending=False)

#set index
gnipc = gnipc.set_index("Country")
gnipc
```

```python
# Transform DF to dict
records_ = gnipc.to_dict(orient = 'records')
# Load data to mongodb
result = db.gross_nat_inc_per_capita.insert_many(records_ )
```
